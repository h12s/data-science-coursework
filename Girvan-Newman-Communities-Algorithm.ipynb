```python
# Part 1 B) Python Part
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community.centrality import girvan_newman
import itertools

edge_list = [('A','B'),('A','C'),('B','C'),('B','H'),('C','D'),('H','I'),
             ('H','G'),('I','G'),('D','E'),('D','F'),('E','F'),('G','E')]
G = nx.Graph(edge_list)

# Stop getting tuples of communities once the number of communities is greater than a 
# suitable number, which in this case is 3 communities, as found in the manual calculation
k = 3
girvan_newman_communities = girvan_newman(G)
limited = itertools.takewhile(lambda c: len(c) <= k, girvan_newman_communities)
for community in limited:
    communities = tuple(sorted(c) for c in community)
    
print("Communities by Girvan-Newman Algorithm")
print("(3 was the suitable number of communities/connected components found in the manual calculation)")
print(communities)

communities = list(communities)
color_map = []
for node in G:
    if node in communities[0]:
        color_map.append("red")
    elif node in communities[1]:
        color_map.append("green")
    elif node in communities[2]:
        color_map.append("orange")
        
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```
